```python
from sklearn.linear_model import LinearRegression

# Re-initialize and fit the model to ensure it's fitted before prediction
modello = LinearRegression()
modello.fit(x_procesata, y_procesata)

```

```python
# Make a prediction for a new Celsius value (e.g., 9 degrees Celsius)
predicted_fahrenheit_12 = modello.predict([[12]])
print(f"Predicted Fahrenheit for 12C: {predicted_fahrenheit_12[0][0]:.2f}F")
```

```python
modello.score(x_procesata,y_procesata)
```
